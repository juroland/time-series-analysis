```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
x = np.arange(-500, 500)/10
n = x.shape[0]
mean_values = 1 / (1 + np.exp(-x))
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(mean_values)
```

```python
values = pd.Series(mean_values + np.random.normal(scale=0.1, size=n))
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(values)
```

```python
rolling_q1 = values.rolling(50).quantile(0.25).fillna(method="backfill")
rolling_median = values.rolling(50).quantile(0.5).fillna(method="backfill")
rolling_q3 = values.rolling(50).quantile(0.75).fillna(method="backfill")
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(rolling_median, label="median")
ax.plot(rolling_q1, label="first quartile")
ax.plot(rolling_q3, label="third quartile")

ax.plot(rolling_q1 > rolling_q3.shift(40).fillna(method="backfill"), label="change point")

ax.legend()
```
